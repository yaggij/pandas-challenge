```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "../Resources/schools_complete.csv"
student_data_to_load = "../Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
schools_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data = pd.merge(student_data, schools_data, how="left", on=["school_name", "school_name"])

school_budget = school_data.mean()
school_data.head()
```

```python
schools = len(school_data["school_name"].unique())
students = school_data["student_name"].value_counts().sum()
budget = school_budget["budget"].mean()
avg_math = school_data["math_score"].mean()
avg_read = school_data["reading_score"].mean()
```

```python
math_pass = school_data[school_data["math_score"]>= 70]["math_score"].count()
math_pct = math_pass/students

read_pass = school_data[school_data["reading_score"]>= 70]["reading_score"].count()
read_pct = read_pass/students

overall_pass = school_data[(school_data['math_score'] >= 70) & (school_data['reading_score'] >= 70)]['student_name'].count()/students

district_summary = pd.DataFrame({
    "Total Schools": [schools],
    "Total Students": [students],
    "Total Budget": [budget],
    "Average Reading Score": [avg_read],
    "Average Math Score": [avg_math],
    "% Passing Reading":[read_pct],
    "% Passing Math": [math_pct],
    "Overall Passing Rate": [overall_pass]})

district_summary
```

```python
grouped = school_data.groupby(["school_name"])

school_types = school_data.set_index("school_name")['type']

students_school = grouped["student_name"].count()
school_budget = school_data.set_index("school_name")["budget"]
budget_per = school_budget/students_school

read_scores_school = grouped["reading_score"].mean()
math_scores_school = grouped["math_score"].mean()

pass_math_group = school_data[school_data['math_score'] >= 70].groupby('school_name')['Student ID'].count()/students_school
pass_read_group = school_data[school_data['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/students_school
overall_group =  school_data[(school_data['math_score'] >= 70) & (school_data['reading_score'] >= 70)].groupby('school_name')['Student ID'].count()/students_school

school_summary = pd.DataFrame({
    "School Type": [school_types],
    "Total Students": [students_school],
    "Total School Budget": [school_budget],
    "Per Student Budget": [budget_per],
    "Average Reading Score": [read_scores_school],
    "Average Math Score": [math_scores_school],
    "% Passing Reading":[pass_read_group],
    "% Passing Math": [pass_math_group],
    "Overall Passing Rate": [overall_group]})

school_summary
```

```python
Genius_schools = school_summary.sort_values("Overall Passing Rate", ascending = False)
Genius_schools
```

```python
Bottom_five = school_summary.sort_values("Overall Passing Rate", ascending = True)
Bottom_five
```

```python
nine_math = school_data.loc[school_data["grade"]== "9th"].groupby("school_name")["math_score"].mean()
ten_math = school_data.loc[school_data["grade"]== "10th"].groupby("school_name")["math_score"].mean()
eleven_math = school_data.loc[school_data["grade"]== "11th"].groupby("school_name")["math_score"].mean()
twelve_math = school_data.loc[school_data["grade"]== "12th"].groupby("school_name")["math_score"].mean()

math_by_grade = pd.DataFrame ({
        "9th": [nine_math],
    "10th": [ten_math],
    "11th": [eleven_math],
    "12th": [twelve_math],
})

math_by_grade
```

```python
nine_read = school_data.loc[school_data["grade"]== "9th"].groupby("school_name")["reading_score"].mean()
ten_read = school_data.loc[school_data["grade"]== "10th"].groupby("school_name")["reading_score"].mean()
eleven_read = school_data.loc[school_data["grade"]== "11th"].groupby("school_name")["reading_score"].mean()
twelve_read = school_data.loc[school_data["grade"]== "12th"].groupby("school_name")["reading_score"].mean()

reading_by_grade = pd.DataFrame ({
        "9th": [nine_read],
    "10th": [ten_read],
    "11th": [eleven_read],
    "12th": [twelve_read],
})

reading_by_grade
```

```python
bins = [0,584.99,614.99,644.99,999999]
group_names = ["< $585", "$585 - 614", "$615 - 644", "> $644"]

school_data["spending"] = pd.cut(school_data['budget']/school_data['size'],bins, labels = group_names)

group_spending = school_data.groupby("spending")

avg_math = group_spending['math_score'].mean()
avg_read = group_spending['reading_score'].mean()
pass_math = school_data[school_data['math_score'] >= 70].groupby('spending')['Student ID'].count()/group_spending['Student ID'].count()
pass_read = school_data[school_data['reading_score'] >= 70].groupby('spending')['Student ID'].count()/group_spending['Student ID'].count()
overall = school_data[(school_data['reading_score'] >= 70) & (school_data['math_score'] >= 70)].groupby('spending')['Student ID'].count()/group_spending['Student ID'].count()

scores_spending = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "Overall Passing Rate": overall
            
})

scores_spending
```

```python
bins = [0,1000,2000,5000]
group_names = ["Small(<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

school_data["sizes"] = pd.cut(school_data['size'],bins, labels = group_names)

group_sizes = school_data.groupby("sizes")

avg_math = group_sizes['math_score'].mean()
avg_read = group_sizes['reading_score'].mean()
pass_math = school_data[school_data['math_score'] >= 70].groupby('sizes')['Student ID'].count()/group_sizes['Student ID'].count()
pass_read = school_data[school_data['reading_score'] >= 70].groupby('sizes')['Student ID'].count()/group_sizes['Student ID'].count()
overall = school_data[(school_data['reading_score'] >= 70) & (school_data['math_score'] >= 70)].groupby('sizes')['Student ID'].count()/group_sizes['Student ID'].count()

scores_spending = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "Overall Passing Rate": overall
            
})

scores_spending
```
